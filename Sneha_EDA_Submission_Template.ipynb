### Variables Description

# 1. Channel Identification & Metadata

 **rank:** The numerical rank of the YouTube channel based on performance.

 **Youtuber:**  The name of the YouTube channel/content creator.

 **category:** The primary content genre (e.g., Music, Entertainment, Gaming).

**created_year / created_month / created_date:** The temporal information indicating when the channel was established.

**Country / Abbreviation:** The country where the channel is primarily based.

# 2. Performance Metrics (Growth & Engagement)

**subscribers:** Total number of subscribers gained by the channel.

**video views:** Total lifetime views across all videos on the channel.

**uploads:** The total number of videos uploaded by the channel.

**subscribers_for_last_30_days:** Recent growth metric showing new subscribers gained in the last month.

**video_views_for_the_last_30_days:** Recent viewership metric showing total views generated in the last month.

# 3. Ranking & Classification

**video_views_rank:** Channel ranking specifically based on total video views.

**country_rank:** The rank of the channel within its respective country.

**channel_type_rank:** The rank of the channel within its content category.

# 4. Financial Metrics (Earnings)

**lowest_monthly_earnings / highest_monthly_earnings:** Estimated range of monthly revenue generated by the channel.

**lowest_yearly_earnings / highest_yearly_earnings:** Estimated range of annual revenue generated by the channel.

# 5. Socio-Economic Indicators (External Data)

**Population:** The total population of the channel's country.

**Unemployment rate:** The percentage of the unemployed population in the country.

**Urban_population:** The total number of people living in urban areas in the country.

**Gross tertiary education enrollment (%):** The percentage of the population enrolled in higher education.

**Latitude / Longitude:** Geographic coordinates of the channel's country.

## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.

# Handling missing values

# Fill categorical columns
df['category'] = df['category'].fillna('Unknown')
df['Country'] = df['Country'].fillna('Unknown')
df['Abbreviation'] = df['Abbreviation'].fillna('Unknown')
df['channel_type'] = df['channel_type'].fillna('Unknown')

# Fill numeric column
df['subscribers_for_last_30_days'] = df['subscribers_for_last_30_days'].fillna(0)

# Drop rows where created_year is missing
df.dropna(subset=['created_year'], inplace=True)

numeric_cols = [
    'country_rank',
    'channel_type_rank',
    'video_views_for_the_last_30_days',
    'Gross tertiary education enrollment (%)',
    'Population',
    'Unemployment rate',
    'Urban_population',
    'Latitude',
    'Longitude'
]

for col in numeric_cols:
    df[col] = df[col].fillna(df[col].median())

print("\nMissing values after cleaning:")
print(df.isnull().sum())

print("Total Missing Values:", df.isnull().sum().sum())

print("\nFinal dataset shape:", df.shape)
````

````python
# Visualizing the missing values
plt.figure(figsize=(14, 8))
sns.heatmap(df.isnull(), cbar=False, cmap='viridis', yticklabels=False)

plt.title('Heatmap of Missing Values (Yellow = Missing Data)')
plt.show()
````
