# 1. Channel Identification & Metadata

 **rank:** The numerical rank of the YouTube channel based on performance.

**lowest_monthly_earnings / highest_monthly_earnings:** Estimated range of monthly revenue generated by the channel.

**lowest_yearly_earnings / highest_yearly_earnings:** Estimated range of annual revenue generated by the channel.

**Population:** The total population of the channel's country.

## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.

# Handling missing values

# Fill categorical columns
df['category'] = df['category'].fillna('Unknown')
df['Country'] = df['Country'].fillna('Unknown')
df['Abbreviation'] = df['Abbreviation'].fillna('Unknown')
df['channel_type'] = df['channel_type'].fillna('Unknown')

# Fill numeric column
df['subscribers_for_last_30_days'] = df['subscribers_for_last_30_days'].fillna(0)

# Drop rows where created_year is missing
df.dropna(subset=['created_year'], inplace=True)

numeric_cols = [
    'country_rank',
    'channel_type_rank',
    'video_views_for_the_last_30_days',
    'Gross tertiary education enrollment (%)',
    'Population',
    'Unemployment rate',
    'Urban_population',
    'Latitude',
    'Longitude'
]

for col in numeric_cols:
    df[col] = df[col].fillna(df[col].median())

print("\nMissing values after cleaning:")
print(df.isnull().sum())

print("Total Missing Values:", df.isnull().sum().sum())

print("\nFinal dataset shape:", df.shape)
````

````python
# Visualizing the missing values
plt.figure(figsize=(14, 8))
sns.heatmap(df.isnull(), cbar=False, cmap='viridis', yticklabels=False)

plt.title('Heatmap of Missing Values (Yellow = Missing Data)')
plt.show()
````

* **Missing Value Imputation:** Handled missing data by filling categorical fields (like category, Country) with 'Unknown' and numerical fields (like subscribers_for_last_30_days) with 0.

* **Data Dropping:** Removed rows with critical missing temporal data (created_year) to ensure accuracy in growth analysis.

* **Data Type Conversion:** Converted created_year to integer format and subscribers to numeric to facilitate statistical calculations and plotting.

* **Deduplication:** Applied drop_duplicates() to ensure the uniqueness of records for reliable analysis.

## ***4. Data Vizualization, Storytelling & Experimenting with charts : Understand the relationships between variables***

#### Chart - 1

````python
# Chart - 1 visualization code

# Chart 1: Distribution of Subscribers

# Styling setup
sns.set_theme(style="whitegrid") # Clean background grid
plt.figure(figsize=(12, 7))

# Data preparation
top_10 = df.nlargest(10, 'subscribers')

# Plotting with a professional color palette
ax = sns.barplot(data=top_10, x='Youtuber', y='subscribers', palette='viridis', hue='Youtuber', legend=False)

# Styling and Labels
plt.title("Top 10 YouTube Channels by Subscribers", fontsize=16, fontweight='bold', pad=20)
plt.xlabel("Channel Name", fontsize=12)
plt.ylabel("Subscribers (in millions)", fontsize=12)
plt.xticks(rotation=45, ha='right')

# Adding value labels on top of bars
for p in ax.patches:
    ax.annotate(f'{p.get_height():.1f}',
                (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center',
                xytext=(0, 9),
                textcoords='offset points', fontsize=10)

plt.tight_layout()
plt.show()
````

##### 1. Why did you pick the specific chart?

A bar chart was chosen because it is the most effective way to compare subscriber counts across different YouTube channels. It clearly highlights the top 10 channels, making it easy to identify which channels have the highest number of subscribers and to compare their performance visually.
