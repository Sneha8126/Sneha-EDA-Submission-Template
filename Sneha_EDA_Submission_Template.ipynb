## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.

# Handling missing values

# Fill categorical columns
df['category'] = df['category'].fillna('Unknown')
df['Country'] = df['Country'].fillna('Unknown')
df['Abbreviation'] = df['Abbreviation'].fillna('Unknown')
df['channel_type'] = df['channel_type'].fillna('Unknown')

# Fill numeric column
df['subscribers_for_last_30_days'] = df['subscribers_for_last_30_days'].fillna(0)

# Drop rows where created_year is missing
df.dropna(subset=['created_year'], inplace=True)

numeric_cols = [
    'country_rank',
    'channel_type_rank',
    'video_views_for_the_last_30_days',
    'Gross tertiary education enrollment (%)',
    'Population',
    'Unemployment rate',
    'Urban_population',
    'Latitude',
    'Longitude'
]

for col in numeric_cols:
    df[col] = df[col].fillna(df[col].median())

print("\nMissing values after cleaning:")
print(df.isnull().sum())

print("Total Missing Values:", df.isnull().sum().sum())

print("\nFinal dataset shape:", df.shape)
````

````python
# Visualizing the missing values
plt.figure(figsize=(14, 8))
sns.heatmap(df.isnull(), cbar=False, cmap='viridis', yticklabels=False)

plt.title('Heatmap of Missing Values (Yellow = Missing Data)')
plt.show()
````

### What all manipulations have you done and insights you found?



# Data Manipulations Performed

* **Missing Value Imputation:** Handled missing data by filling categorical fields (like category, Country) with 'Unknown' and numerical fields (like subscribers_for_last_30_days) with 0.

* **Data Dropping:** Removed rows with critical missing temporal data (created_year) to ensure accuracy in growth analysis.

* **Data Type Conversion:** Converted created_year to integer format and subscribers to numeric to facilitate statistical calculations and plotting.

* **Deduplication:** Applied drop_duplicates() to ensure the uniqueness of records for reliable analysis.

## Chart - 16

````python
# Data preparation
channel_counts = df['channel_type'].value_counts()

# Plotting
fig, ax = plt.subplots(figsize=(12, 7))
colors = plt.cm.Set3.colors

# Function to hide labels for very small slices (< 3%)
def my_autopct(pct):
    return ('%1.1f%%' % pct) if pct > 3 else ''

# Donut Chart with conditional labeling
wedges, texts, autotexts = ax.pie(channel_counts, labels=None, autopct=my_autopct,
                                   startangle=140, colors=colors, pctdistance=0.8,
                                   wedgeprops={'width': 0.4, 'edgecolor': 'w', 'linewidth': 2})

# Percentages ka font size adjust karein
for autotext in autotexts:
    autotext.set_fontsize(9)
    autotext.set_fontweight('bold')

# Legend side mein (labels ke liye)
ax.legend(wedges, channel_counts.index, title="Channel Types",
           loc="center left", bbox_to_anchor=(1, 0, 0.5, 1), fontsize=10)

# Center circle
centre_circle = plt.Circle((0,0), 0.60, fc='white')
ax.add_artist(centre_circle)

plt.title('Distribution of Channel Types', fontsize=16, fontweight='bold')
plt.tight_layout()
plt.show()
````

##### 1. Why did you pick the specific chart?

* A donut chart was chosen because it clearly shows the proportion of different channel types in the dataset.

* It is effective for visualizing percentage distribution across categories.

##### 2. What is/are the insight(s) found from the chart?

* Entertainment is the most common channel type, contributing the largest share of channels.

* Music is the second-largest channel type, followed by People and Games.

* Several channel types, such as Animals, Nonprofit, and Autos, represent only a small percentage of the dataset.

* The distribution shows that entertainment-related content dominates YouTube.
