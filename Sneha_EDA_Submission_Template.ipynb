### Variables Description

# 1. Channel Identification & Metadata

 **rank:** The numerical rank of the YouTube channel based on performance.

**subscribers:** Total number of subscribers gained by the channel.

**lowest_monthly_earnings / highest_monthly_earnings:** Estimated range of monthly revenue generated by the channel.

**lowest_yearly_earnings / highest_yearly_earnings:** Estimated range of annual revenue generated by the channel.

# 5. Socio-Economic Indicators (External Data)

**Population:** The total population of the channel's country.

**Unemployment rate:** The percentage of the unemployed population in the country.

**Urban_population:** The total number of people living in urban areas in the country.

## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.

# Handling missing values

# Fill categorical columns
df['category'] = df['category'].fillna('Unknown')
df['Country'] = df['Country'].fillna('Unknown')
df['Abbreviation'] = df['Abbreviation'].fillna('Unknown')
df['channel_type'] = df['channel_type'].fillna('Unknown')

# Fill numeric column
df['subscribers_for_last_30_days'] = df['subscribers_for_last_30_days'].fillna(0)

# Drop rows where created_year is missing
df.dropna(subset=['created_year'], inplace=True)

numeric_cols = [
    'country_rank',
    'channel_type_rank',
    'video_views_for_the_last_30_days',
    'Gross tertiary education enrollment (%)',
    'Population',
    'Unemployment rate',
    'Urban_population',
    'Latitude',
    'Longitude'
]

for col in numeric_cols:
    df[col] = df[col].fillna(df[col].median())

print("\nMissing values after cleaning:")
print(df.isnull().sum())

print("Total Missing Values:", df.isnull().sum().sum())

print("\nFinal dataset shape:", df.shape)
````

````python
# Visualizing the missing values
plt.figure(figsize=(14, 8))
sns.heatmap(df.isnull(), cbar=False, cmap='viridis', yticklabels=False)

plt.title('Heatmap of Missing Values (Yellow = Missing Data)')
plt.show()
````

* **Data Type Conversion:** Converted created_year to integer format and subscribers to numeric to facilitate statistical calculations and plotting.

#### Chart - 1

````python
# Chart - 1 visualization code

# Chart 1: Distribution of Subscribers

# Styling setup
sns.set_theme(style="whitegrid") # Clean background grid
plt.figure(figsize=(12, 7))

# Data preparation
top_10 = df.nlargest(10, 'subscribers')

# Plotting with a professional color palette
ax = sns.barplot(data=top_10, x='Youtuber', y='subscribers', palette='viridis', hue='Youtuber', legend=False)

# Styling and Labels
plt.title("Top 10 YouTube Channels by Subscribers", fontsize=16, fontweight='bold', pad=20)
plt.xlabel("Channel Name", fontsize=12)
plt.ylabel("Subscribers (in millions)", fontsize=12)
plt.xticks(rotation=45, ha='right')

# Adding value labels on top of bars
for p in ax.patches:
    ax.annotate(f'{p.get_height():.1f}',
                (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center',
                xytext=(0, 9),
                textcoords='offset points', fontsize=10)

plt.tight_layout()
plt.show()
````

````python
# Chart - 8 visualization code
import matplotlib.pyplot as plt
import seaborn as sns

# Average subscribers by category
avg_subs = (
    df.groupby('category', as_index=False)['subscribers']
      .mean()
      .sort_values(by='subscribers', ascending=False)
)

sns.set_theme(style="whitegrid")
plt.figure(figsize=(14,7))

ax = sns.barplot(
    data=avg_subs,
    x='category',
    y='subscribers',
    palette='viridis',
    hue='category',
    legend=False
)

plt.title("Average Subscribers by Category", fontsize=16, fontweight='bold')
plt.xlabel("Category")
plt.ylabel("Average Subscribers")
plt.xticks(rotation=45, ha='right')

plt.tight_layout()
plt.show()
````

## Chart - 14

````python
# Data preparation
channel_counts = df['channel_type'].value_counts()

# Plotting
fig, ax = plt.subplots(figsize=(12, 7))
colors = plt.cm.Set3.colors

# Function to hide labels for very small slices (< 3%)
def my_autopct(pct):
    return ('%1.1f%%' % pct) if pct > 3 else ''

# Donut Chart with conditional labeling
wedges, texts, autotexts = ax.pie(channel_counts, labels=None, autopct=my_autopct,
                                   startangle=140, colors=colors, pctdistance=0.8,
                                   wedgeprops={'width': 0.4, 'edgecolor': 'w', 'linewidth': 2})

# Percentages ka font size adjust karein
for autotext in autotexts:
    autotext.set_fontsize(9)
    autotext.set_fontweight('bold')

# Legend side mein (labels ke liye)
ax.legend(wedges, channel_counts.index, title="Channel Types",
           loc="center left", bbox_to_anchor=(1, 0, 0.5, 1), fontsize=10)

# Center circle
centre_circle = plt.Circle((0,0), 0.60, fc='white')
ax.add_artist(centre_circle)

plt.title('Chart 14: Distribution of Channel Types', fontsize=16, fontweight='bold')
plt.tight_layout()
plt.show()

````
