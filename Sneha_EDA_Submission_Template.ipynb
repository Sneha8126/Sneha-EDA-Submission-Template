````python
# Import Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
````

### Dataset Loading

````python
# Load Dataset
#df = pd.read_csv('/content/Global YouTube Statistics.csv',encoding='utf-8',encoding_errors='ignore')
df = pd.read_csv('Global YouTube Statistics.csv', encoding='latin1')
````

### Dataset First View

### Dataset Rows & Columns count

````python
df.columns = df.columns.str.strip()
````

### Dataset Information

````python
# Dataset Info
df.info()
````

#### Duplicate Values

````python
# Dataset Duplicate Value Count
duplicate_count = df.duplicated().sum()
print(f"Number of duplicate rows: {duplicate_count}")
````

#### Missing Values/Null Values

````python
# Missing Values/Null Values Count
print("\nMissing values before cleaning:")
print(df.isnull().sum())
````

````python
df = df.copy()
````

````python
# Visualizing the missing values
plt.figure(figsize=(14, 8))
# df.isnull() humein True/False ka matrix deta hai
sns.heatmap(df.isnull(), cbar=False, cmap='viridis', yticklabels=False)

plt.title('Heatmap of Missing Values (Yellow = Missing Data)')
plt.show()
````

````python
# Handling missing values

# Fill categorical columns
df['category'] = df['category'].fillna('Unknown')
df['Country'] = df['Country'].fillna('Unknown')
df['Abbreviation'] = df['Abbreviation'].fillna('Unknown')
df['channel_type'] = df['channel_type'].fillna('Unknown')

# Fill numeric column
df['subscribers_for_last_30_days'] = df['subscribers_for_last_30_days'].fillna(0)

# Drop rows where created_year is missing
df.dropna(subset=['created_year'], inplace=True)

numeric_cols = [
    'country_rank',
    'channel_type_rank',
    'video_views_for_the_last_30_days',
    'Gross tertiary education enrollment (%)',
    'Population',
    'Unemployment rate',
    'Urban_population',
    'Latitude',
    'Longitude'
]

for col in numeric_cols:
    df[col] = df[col].fillna(df[col].median())

print("\nMissing values after cleaning:")
print(df.isnull().sum())

print("Total Missing Values:", df.isnull().sum().sum())

print("\nFinal dataset shape:", df.shape)
````
